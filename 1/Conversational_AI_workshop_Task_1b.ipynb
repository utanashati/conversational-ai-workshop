# Load model

Here we will download the model weights for GPT-2 using the visualization library ecco.

GPT-2 is pretrained on huge amounts of raw text from books, Wikipedia and Reddit.

You are free to explore other models, as listed [here](https://huggingface.co/models).

```python
# This command downloads the GPT-2 model from Huggingface
lm = ecco.from_pretrained('gpt2', activations=True)
```

# Generate new text using the pretrained model

```python
text = "The important thing is to never stop" #@param {type:"string"}

# Generate one token
output = lm.generate(text, generate=1, do_sample=False)
```

Here we disable sampling using `do_sample=False`. This means we will always pick the most probable token (also known as greedy sampling). Let's see how this probability distribution over tokens looks like.

```python
output.layer_predictions(position=7, layer=11, topk=20)
```

As you can see there are many similarly probable tokens here. Let's see how adding stochasticity by sampling changes the generated text.

```python
for i in range(5):
  lm.generate(text, generate=20, do_sample=True)
```

# Visualizing token importance: Saliency 🌈

Using the saliency command, we can get a heat-map of which other previous tokens were most important in generating a given token. 

In particular, saliency is calculated when backpropagating a given (most probable) token back to the input tokens. This then gives a measure of importance to each token: If a small change in one of the input tokens has a large effect on the predicted token, we can assume that this input token has a high importance for predicting the final token.

```python
text = 'Here is a list of countries: Switzerland,' #@param {type:"string"}

output = lm.generate(text, generate=30, do_sample=False)
```

```python
output.saliency()
```

Notice how attention is usually on countries, where punctuations (such as , or :) are almost ignored.

```python
text = 'Here we are just counting up: 1, 2, 3, 4' #@param {type:"string"}

output = lm.generate(text, generate=40, do_sample=False)
```

```python
output.saliency()
```

We can also follow the ranking of a token throughout the hidden states. In this example the `,` is very easy to predict and ranks first almost from the beginning (layer 0). In the case of the numbers, the ranking gradually increases throughout the layers.

```python
output.rankings()
```

## Neuron activations ⚡️

An interesting feature in some deep learning architectures (such as CNNs) is a certain degree of specialization of subcomponents. We can observe the same thing for transformers.

In a forward pass we pool all layer activations together into one big matrix on which we run [non-negative matrix factorization](https://en.wikipedia.org/wiki/Non-negative_matrix_factorization), which, similar to a PCA, reduces the dimensionality of the matrix into `n` components of our choice.


image.png

```python
text = """Our tutorial focuses on the topic of conversational AI, specifically how state-of-the-art machine learning models for natural language processing work, how they can be used for text generation, and how to employ text generation techniques to build a conversational system. While we will introduce the concepts and theoretical background used for conversational AI, we will not give a complete introduction to natural language processing. Our emphasis is on acquiring skills that allow you to implement a real-world conversational system.

Arguably one of the key drivers for the adoption of the internet has been interaction with other humans, be it through email, messaging platforms or social media. Interactions with computers, such as search engines, have become much easier compared to the early days of the internet. Chatbots and dialogue systems are becoming increasingly integrated into modern online platforms. However, even today, conversations are often clumsy and inefficient, especially when the conversations go beyond a simple Siri command. And although for most people knowing “how to google" and navigating the complex user interfaces of websites is now part of their core digital skillset, there is an abundance of people with visual or other impairments, who greatly struggle with such tasks. Improving conversational AI is therefore an important direction also for improving accessibility and democratisation of the web.

Passing the Turing test has been considered as the holy grail for artificial intelligence. It has also drawn a lot of criticism and its relevance has been questioned by both philosophers and AI researchers. Undoubtedly, the strengths of this test lies in its simplicity and the pragmatic approach to very complex philosophical questions. Our workshop will not aim to give a resolution to these questions, but it might be able to give an idea for how far conversational AI has come, and exactly what we as humans seek out for in conversations with computers.
"""
output = lm.generate(text, generate=1, do_sample=True)
nmf_1 = output.run_nmf(n_components=8) 
```

```python
nmf_1.explore()
```

```python
nmf_1 = output.run_nmf(n_components=8, from_layer=0, to_layer=1)
nmf_1.explore()
```

```python
text = 'Here we are just counting up: 1, 2, 3, 4' #@param {type:"string"}

output = lm.generate(text, generate=40, do_sample=False)
nmf_1 = output.run_nmf(n_components=8)
```

```python
nmf_1.explore()
```
